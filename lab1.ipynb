### Задание №3
Реализовать методы сортировки строк числовой матрицы в соответствии с
заданием. Оценить время работы каждого алгоритма сортировки и сравнить его со
временем стандартной функции сортировки. Испытания проводить на сгенерированных
матрицах.

```python
import copy
import random
import time
user_m = int(input('1st dimension = '))
user_n = int(input('2nd dimension = '))
user_min_limit = int(input('min = '))
user_max_limit = int(input('max = '))
mass=[]
for i in range(user_m):
    mass.append([])
    for j in range(user_n):
        mass[i].append('')
for i in range(user_m):
    for j in range(user_n):
        mass[i][j] = random.randint(user_min_limit, user_max_limit)
```

```python
# Сортировка выбором.
# Берётся срез массива, в котором минимальный элемент переносят в самый левый угол,
# после чего срез уменьшается и цикл повторяется.
arrr=copy.deepcopy(mass)

print(arrr)
   

def vybor(arrr):
 

    i = 0
    N=len(arr)

    while i < N - 1:
        m = i
        j = i + 1
        while j < N:
            if arr[j] < arr[m]:
                m = j
            j += 1
        
        arr[i], arr[m] = arr[m], arr[i]
        i += 1

    return arrr
        
#print(arrr)

start_time = time.time()

for i in range(len(arrr)):
    arr=arrr[i]
    vybor(arr)
    arrr[i]=arr

print(arrr)

print("сортировка выбором = ""--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка вставкой.
arrr=copy.deepcopy(mass)
print(arrr)
def vstav(arrr):
    for i in range(1, len(arr)):
        j = i
        while j > 0 and arr[j-1]>arr[j]:
            arr[j-1], arr[j] = arr[j], arr[j-1]
            j -= 1
    return arrr

start_time = time.time()

for i in range(len(arrr)):
    arr=arrr[i]
    vstav(arr)
    arrr[i]=arr
print(arrr)
print("Сортировка вставкой = ""--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка обменом. || Сортировка пузырьком.

arrr=copy.deepcopy(mass)

print(arrr)

def obmen(arrr):
    N = len(arr)
    i = 0
    while i < N - 1:
        j = 0
        while j < N - 1 - i:
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
            j += 1
        i += 1
    return arrr

start_time = time.time()

for i in range(len(arrr)):
    arr=arrr[i]
    obmen(arr)
    arrr[i]=arr

print(arrr)

print("сортировка вставкой = ""--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка Шелла.
arrr=copy.deepcopy(mass)

print(arrr)

def shell(arrr):
    

    length = len(arr)
    gap = length//2
    while gap > 0:
        for iIndex in range(gap, length):
            temp = arr[iIndex]
            jIndex = iIndex
            while jIndex >= gap and arr[jIndex - gap] > temp:
                arr[jIndex] = arr[jIndex - gap]
                jIndex -= gap
            arr[jIndex] = temp
        gap //= 2
    return arrr

start_time = time.time()

for i in range(len(arrr)):
    arr=arrr[i]
    shell(arr)
    arrr[i]=arr


print(arrr)

print("сортировка шелла = ""--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Быстрая сортировка.
arrr=copy.deepcopy(mass)

print(arrr)

def partition(arr, low, high):
    i = (low - 1)         # index of smaller element
    pivot = arr[high]     # pivot

    for j in range(low, high):
        # If current element is smaller than the pivot
        if arr[j] < pivot:

            # increment index of smaller element
            i = i + 1
            arr[i], arr[j] = arr[j], arr[i]

    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1

# The main function that implements QuickSort
# arr[] --> Array to be sorted,
# low  --> Starting index,
# high  --> Ending index

# Function to do Quick sort
def quick_sort(arr, low, high):
    if low < high:
        # pi is partitioning index, arr[p] is now
        # at right place
        pi = partition(arr, low, high)

        # Separately sort elements before
        # partition and after partition
        quick_sort(arr, low, pi - 1)
        quick_sort(arr, pi + 1, high)
    
start_time = time.time()

for i in range(len(arrr)):
    arr=arrr[i]

    quick_sort(arr,0,len(arr)-1)
  
    arrr[i]=arr
 
print(arrr)      
   
print("быстрая сортировка = ""--- {0} ms ---".format(round((time.time() - start_time)*1000))) 



 
```

```python
# Турнирная сортировка.

arrr=copy.deepcopy(mass)

print(arrr)



def heapify(arr, n, root):
    largest = root
    left = 2 * root + 1
    right = 2 * root + 2

    # See if left child of root exists and is greater than root
    if left < n and arr[root] < arr[left]:
        largest = left
    # See if right child of root exists and is greater than root
    if right < n and arr[largest] < arr[right]:
        largest = right

    # Change root, if needed
    if largest != root:
        arr[root], arr[largest] = arr[largest], arr[root]

        # Heapify the root.
        heapify(arr, n, largest)


def heap_sort(arr):
    n = len(arr)

    # Build a maxheap.
    for i in range(n, -1, -1):
        heapify(arr, n, i)

    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        heapify(arr, i, 0)






start_time = time.time()

for i in range(len(arrr)):
    arr=arrr[i]
 
    heap_sort(arr)
   
    arrr[i]=arr
 
      

print(arrr)

print("турнирная сортировка = ""--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```
